# lrdata to MuData

### Load packages

```python
from pathlib import Path
import scanpy as sc
import liana as li
```

### Convert lrdata to muData

```python
lrdata_path  = Path("/g/stegle/aalsayah/repos/data/merfish")
adata = sc.read_h5ad(lrdata_path / "lrdata_animal2_C57BL6J-2.039.h5ad")
```

```python
mdata = li.utils.anndata_to_mdata(
    adata = adata,
    groupby = "cell_type",
    expr_prop = 0,
    by_var = False,
    verbose = True)
```

### Filter out modalities if ncells < 10 and/or ngenes <10

```python
for mod_name, adata in mdata.mod.items():
    cells, genes = adata.shape  
    print(f"Modality: {mod_name}, Cells: {cells}, Genes: {genes}")
```

```python
filtered_mods = {key: adata for key, adata in mdata.mod.items() if adata.shape[0] >= 10 and adata.shape[1] >= 10}
mdata.mod = filtered_mods
```

```python
for mod_name, adata in mdata.mod.items():
    cells, genes = adata.shape  
    print(f"Modality: {mod_name}, Cells: {cells}, Genes: {genes}")
```

### Save mdata

```python
mdata.write_h5mu(lrdata_path / "mdata_merfish_brain.h5mu")
```
